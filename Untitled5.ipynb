```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
import seaborn as sns
import random
```

```python
df = pd.read_csv("train.csv")
```

```python
#df['total_qty_over_index'] = df['total_qty_over_index'].apply(lambda x: str(x))
#df['total_qty_over_index_and_type'] = df['total_qty_over_index_and_type'].apply(lambda x: str(x))
#df['total_qty_oper_login_0'] = df['total_qty_oper_login_0'].apply(lambda x: str(x))
#df['total_qty_oper_login_1'] = df['total_qty_oper_login_1'].apply(lambda x: str(x))
#df['dist_qty_oper_login_1'] = df['dist_qty_oper_login_1'].apply(lambda x: str(x))

df['index_oper'] = df['index_oper'].apply(lambda x: str(x))
df['priority'] = df['priority'].apply(lambda x: str(x))
df['class'] = df['class'].apply(lambda x: str(x))
df['mailtype'] = df['mailtype'].apply(lambda x: str(x))
df['mailctg'] = df['mailctg'].apply(lambda x: str(x))
df['directctg'] = df['directctg'].apply(lambda x: str(x))
df['postmark'] = df['postmark'].apply(lambda x: str(x))
df['is_wrong_sndr_name'] = df['is_wrong_sndr_name'].apply(lambda x: str(x))
df['is_wrong_rcpn_name'] = df['is_wrong_rcpn_name'].apply(lambda x: str(x))
df['is_wrong_phone_number'] = df['is_wrong_phone_number'].apply(lambda x: str(x))
df['is_wrong_address'] = df['is_wrong_address'].apply(lambda x: str(x))
df['label'] = df['label'].apply(lambda x: str(x))
```

```python
del df['oper_type + oper_attr']
del df['id']
del df['mailrank']
```

```python
features = ["weight", "transport_pay", "weight_mfi", "price_mfi", 
              "total_qty_over_index", "total_qty_over_index_and_type", "total_qty_oper_login_0",
           "total_qty_oper_login_1", "dist_qty_oper_login_1"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(df[features].to_numpy())
k = 0
for i in features:
    df[i] = X[:,k]
    k+=1
```

```python
X = ""
```

```python
del df['name_mfi']
```

```python
features.append("label")
category = []
for i in df.columns:
    if i not in features: category.append(i)
```

```python
print(category)
```

```python
from category_encoders.binary import BinaryEncoder
d = pd.DataFrame()
ct = []
binens = []
for i in category:
    if i != "name_mfi":
        print(i)
        bn = BinaryEncoder()
        bn.fit(df[i].values)
        binens.append(bn)
        k = bn.transform(df[i].values)
        l  = []
        for j in k.columns:
            l.append(str(i)+str(j))
        ct.extend(l)
        k.columns = l
        d = pd.concat([d, k], axis=1)
```

```python
np.random.seed(77)
val_ds = pd.DataFrame()
train_ds = pd.DataFrame()
for label in df.label.unique():
    n = 75000
    ids = random.sample(list(df.loc[df.label == label].index), n)
    val_ds = pd.concat([val_ds, df.loc[ids]])
    train_ids = list(set(df.loc[df.label == label].index) - set(ids))
    train_ds = pd.concat([train_ds, df.loc[train_ids]])
```

```python
val_ds.shape[0]/train_ds.shape[0]
```

```python
n = train_ds.loc[train_ds["label"]=='1'].shape[0]
ids = random.sample(list(train_ds.loc[train_ds["label"]=='0'].index), n)
train_ds = pd.concat([train_ds.loc[train_ds["label"]=='1'], train_ds.loc[ids]])
```

```python
train_ds.shape
```

```python
train = train_ds.copy()
y = train['label']
del train['label']
X = train
```

```python
catmodel = CatBoostClassifier(iterations=1000,
                              cat_features = ct,
                             task_type = 'GPU')
```

```python
catmodel.fit(X, y)
```

```python
# посмотрим на важность признаков
imp = pd.DataFrame({
    'feature_name':X.columns,
    'feature_score': catmodel.feature_importances_
}).sort_values('feature_score', ascending=False).reset_index(drop=True)

imp.to_numpy()
```

```python
from sklearn.metrics import recall_score
X_test = val_ds.copy()
del X_test['label']
pred = catmodel.predict(X_test)
#X = X_test['postmark'].copy()
#X.loc[X==0.0] = 0
#X.loc[X!=0] = 1
#pred = pr(X.to_list())
print(catmodel,'\n',recall_score(val_ds.label.to_list(), pred, average="macro", zero_division=0))
```
